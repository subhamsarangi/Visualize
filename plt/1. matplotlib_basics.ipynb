```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
plt.style.use('seaborn')
```

```python
import numpy as np
```

```python
x = np.linspace(15,70,80)
y = np.tan(x)
```

## Basic method

```python
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('tan(Y)')
```

#### Multiple plots at once

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'g')
```

## Object oriented method
_We can do the same things using Object oriented method_

```python
fig=plt.figure()
axes=fig.add_axes([0,0,.8,.8])
axes.set(xlabel='x',ylabel='tan(y)')
axes.plot(x,y)
```

```python
fig=plt.figure()
ax1=fig.add_axes([0,0,.5,1])
ax2=fig.add_axes([0.55,0,.5,1])
ax1.plot(x,y,'r')
ax2.plot(y,x,'g')
```

## Plotting Inside another plot

```python
fig = plt.figure()
ax1 = fig.add_axes([.1,.1,.8,.8])
ax2 = fig.add_axes([.2,.2,.4,.3])
ax2.set(xlabel='INSET',xticks=[],ylim=[30,50])
ax1.set(xlabel='OUTER')
ax1.plot(x,y,'g')
ax2.plot(y,x)
```

### Subplots
_Tuple unpacking_

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
for i in range(len(axes)):
    axes[i].plot(x,y**(i+1))
```

### Decoration

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(np.arange(50),np.arange(50)**2,
       lw=2,marker='o',markevery=10,markersize=20,
       markerfacecolor='#372d3b',markeredgewidth=4,
       markeredgecolor='#f1c131')
```
