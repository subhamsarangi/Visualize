```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
import numpy as np
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

## Barplot
__x__ is a _catogorical_ column and __y__ is a _numerical_ column and the default __estimator__ function is _mean_

```python
sns.barplot(x='sex',y='total_bill',data=tips)
plt.savefig('images/sns_7barplot.png',dpi=200)
```

We can give any estimator we want

```python
sns.barplot(x='day',y='tip',data=tips,estimator=np.std)
plt.savefig('images/sns_7barplot_est.png',dpi=200)
```

## Count plot
the number of occurences

```python
sns.countplot(x='sex',data=tips)
plt.savefig('images/sns_8countplot.png',dpi=200)
```

## Box Plots
__Shows the quatiles of the dataset__ _while the whiskers extend to the rest of the distribution._

The distribution of quantitative data in a way that, hopefully, facilitates comparisions between variables

```python
sns.boxplot(x='day',y='total_bill',data=tips)
plt.savefig('images/sns_9boxplot.png',dpi=200)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='sex')
plt.savefig('images/sns_9boxplot_hue.png',dpi=200)
```

## Violin Blot
a type of kdeplot

```python
sns.violinplot(x='day',y='total_bill',data=tips)
plt.savefig('images/sns_10violinplot.png',dpi=200)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')
plt.savefig('images/sns_10violinplot_hue.png',dpi=200)
```

The `split` attribute divides a violin according to the categorical data which is given in the `hue` attribute.

__NOTE__: _There must be exactly two hue levels to use _`split`.

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
plt.savefig('images/sns_10violinplot_split.png',dpi=200)
```

## Strip Plot

```python
sns.stripplot(x='day',y='total_bill',data=tips)
plt.savefig('images/sns_11stripplot.png',dpi=200)
```

The `jitter` attribute adds random noise to separate the stacked points.

We can also use the `hue` attribute and the `split` attribute here.


```python
sns.stripplot(x='day',y='tip',data=tips,jitter=True)
plt.savefig('images/sns_11stripplot_jitter.png',dpi=200)
```

## Swarmplot
combines a stripplot with a violinplot

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
plt.savefig('images/sns_12swarmplot.png',dpi=200)
```

_It's a good idea to combine a swarm plot with a violin plot_

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='#eeeeee')
plt.savefig('images/sns_12swarm_and_violin.png',dpi=200)
```

## Factorplot
The most general form of all these plot. It takes a `kind` argument, where we can specify whether it's bar, box, violin, strip, swarm.

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='bar')
```
