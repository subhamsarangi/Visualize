# Regression Plots

```python
import seaborn as sns
import matplotlib.pyplot as plt
% matplotlib inline
sns.set_style('darkgrid')
```

We load the tips dataset from seaborn

```python
tips = sns.load_dataset('tips')
tips.head()
```

## Linear Model Plot

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='magma')
plt.savefig('images/sns_16lmplot.png',dpi=200)
```

Separate the plot by column and/or rows

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='time')
plt.savefig('images/sns_16lmplot_colrow.png',dpi=200)
```

Customizing the plot using matplotlib

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'],
          scatter_kws={'s':100})
plt.savefig('images/sns_16lmplot_2.png',dpi=200)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',aspect=0.6,size=8)
plt.savefig('images/sns_16lmplot_3.png',dpi=200)
```
