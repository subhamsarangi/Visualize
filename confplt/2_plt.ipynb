```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
style.use('classic')
%matplotlib inline
```

```python
np.random.seed(1)
x = np.arange(5)
y = np.random.randn(5)

fig, axes = plt.subplots(ncols=2, figsize=plt.figaspect(1./2))

vert_bars = axes[0].bar(x, y, color='#372d3b', align='center')
horiz_bars = axes[1].barh(x, y, color='#f1c131', align='center')

# I'll also introduce axhline & axvline to draw a line all the way across the axes
# This can be a quick-n-easy way to draw an axis "spine".
axes[0].axhline(0, color='gray', linewidth=2)
axes[1].axvline(0, color='gray', linewidth=2)
plt.title(fig)
fig.savefig('plot_example_1.png', facecolor='#dddddd',dpi=200)
```

```python
fig, ax = plt.subplots()
vert_bars = ax.bar(x, y, color='indianred', align='center')

# We could have also done this with two separate calls to `ax.bar` and numpy boolean indexing.
for bar, height in zip(vert_bars, y):
    if height < 0:
        bar.set(edgecolor='#222222', color='royalblue', linewidth=3)

plt.title(fig)
fig.savefig('plot_example_2.png', facecolor='#dddddd',dpi=200)
```

```python
x = np.linspace(0, 10, 100)

fig, axes = plt.subplots(ncols=3, figsize=(15,4))
for ax in axes:
    ax.margins(y=0.10)

# Default plotting, colors will be determined by the axes' color_cycle
for i in range(1, 6):
    axes[0].plot(x, i * x,lw=2)

# Demonstrating different linestyles
for i, ls in enumerate(['-', '--', ':', '-.']):
    axes[1].plot(x, np.cos(x) + i, linestyle=ls,lw=2)

# Using linestyles and markers
for i, (ls, mk) in enumerate(zip(['', '-', ':'], ['o', '^', 's'])):
    axes[2].plot(x, np.cos(x) + i * x, linestyle=ls, marker=mk, markevery=10,lw=2)

plt.title(fig)
fig.savefig('plot_example_3.png', facecolor='#dddddd',dpi=250)
```

```python
np.random.seed(1874)
x, y, z = np.random.normal(0, 1, (3, 100))
t = np.arctan2(y, x)
size = 50 * np.cos(2 * t)**2 + 10
```

```python
fig, axes = plt.subplots(ncols=3, figsize=(15,6))
axes[0].scatter(x, y, marker='o', facecolor='white', s=80, edgecolor='black')

axes[1].scatter(x, y, s=size, marker='s', color='darkblue')

axes[2].scatter(x, y, c=z, s=size, cmap='gist_ncar',edgecolor='black')

plt.title(fig,y=0.95)
fig.savefig('scatter_example.png', facecolor='#dddddd',dpi=300)
```

```python
def main():
    fig, axes = plt.subplots(ncols=3, figsize=(15,6))
    plt.title('"ax.bar(...)": Plot rectangles')
    basic_bar(axes[0])
    tornado(axes[1])
    general(axes[2])
    fig.savefig('bar_example.png', facecolor='#dddddd',dpi=300)
    plt.show()

def basic_bar(ax):
    y = [1, 3, 4, 5.5, 3, 2]
    err = [0.2, 1, 2.5, 1, 1, 0.5]
    x = np.arange(len(y))
    ax.bar(x, y, yerr=err, color='#00ab6b', ecolor='#333333')
    ax.margins(0.05)
    ax.set_ylim(bottom=0)
    ax.set_label('bar(x, y, yerr=e)')

def tornado(ax):
    y = np.arange(8)
    x1 = y + np.random.random(8) + 1
    x2 = y + 3 * np.random.random(8) + 1
    ax.barh(y, x1, color='indianred')
    ax.barh(y, -x2, color='#372d3b')
    ax.margins(0.15)
    ax.set_label('barh(x, y)')

def general(ax):
    num = 10
    left = np.random.randint(0, 10, num)
    bottom = np.random.randint(0, 10, num)
    width = np.random.random(num) + 0.5
    height = np.random.random(num) + 0.5
    ax.bar(left, height, width, bottom, color='#f1c131')
    ax.margins(0.15)
    ax.set_label('bar(l, h, w, b)')

main()
```
